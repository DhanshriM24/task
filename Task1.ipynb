```python
import pandas as pd
```

#### 1.Downloaded the excel files<br>
##### 2.Read the data files

```python
#reading data1
df1 = pd.read_csv('data1.csv')
df1.head()
```

```python
#reading data2
df2 = pd.read_csv('data2.csv')
df2
```

#### 3.Remove space between Product ID and Order ID by renaming the column as Product_ID and Order_ID respectively

```python
#renaming columns
df1.rename(columns={'Order ID': 'Order_ID', 'Product ID': 'Product_ID'}, inplace=True)
df2.rename(columns={'Order ID': 'Order_ID', 'Product ID': 'Product_ID'}, inplace=True)
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1["New_ID"] = df1["Order_ID"]+'_'+df1["Product_ID"]
df2["New_ID"] = df2["Order_ID"]+'_'+df2["Product_ID"]
```

```python
df1.head()
```

```python
df2.head()
```

1. **How to identify the Records (Order ID + Product ID combination) present in data1 but missing in data2**

    ***(Specify the number of records missing in your answer*)**

```python
#1st way

data1=set(df1["New_ID"]).difference(set(df2["New_ID"]))
```

```python
#no of records present in data1 but missing in data2
pd.DataFrame(data1).value_counts().sum()
```

```python
#2nd way

x = df1[df1.New_ID.isin(df2.New_ID) == False]
```

```python
x
```

```python
#no of records present in data1 but missing in data2
x.shape[0]
```

2. **How to identify the Records (Order ID + Product ID combination) present in data2 but missing in data1**

  ***(Specify the number of records missing in your answer)***

```python
#1st way

data2=set(df2["New_ID"]).difference(set(df1["New_ID"]))
```

```python
#no of records present in data2 but missing in data1
pd.DataFrame(data2).value_counts().sum()
```

```python
#2nd way

y = df2[df2.New_ID.isin(df1.New_ID) == False]
```

```python
y
```

```python
#no of records present in data2 but missing in data1
y.shape[0]
```

3. **Find the Sum of the total Qty of Records missing in data1 but present in data2** 

```python
#'y' dataframe gives number of records present in data2 but not in data 1 then calculate sum of qty of 'y' dataframe 
print(y['Qty'].sum())
```

4. **Find the total number of unique records (Order ID  Product ID combination) present in the combined dataset of data1 and data2**

```python
#combine dataframe x & y

combined_xy = pd.concat([x,y])
```

```python
combined_xy
```

```python
#intersection of df1 and df2
inter_df = df1.merge(df2)
```

```python
#appending 'inter_df' and 'combined_xy'
unique_df = pd.concat([inter_df, combined_xy])
```
